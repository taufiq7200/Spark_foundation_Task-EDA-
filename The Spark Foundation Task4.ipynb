```python
filterYear = data['iyear']  ==1970
```

```python
filterData = data[filterYear]

reqfilterData = filterData.loc[: ,'city':'longitude']
reqfilterData = reqfilterData.dropna()
reqfilterDataList = reqfilterData.values.tolist()
```

```python
killData = data.loc[:,'nkill']
print('Number of people killed by terror attck:' , int(sum(killData.dropna())))
```

```python
attack_data = data.loc[:,'attacktype1':'attacktype1_txt']

type_killdata = pd.concat([attack_data , killData], axis=1)
```

```python
type_kill_format_data = type_killdata.pivot_table(columns='attacktype1_txt' , values='nkill' , aggfunc='sum')
type_kill_format_data
```

```python
type_kill_format_data.info()
```

```python
labels = type_kill_format_data.columns.tolist()
transpose = type_kill_format_data.T
values= transpose.values.tolist()
fig , ax = plt.subplots(figsize=(20,20) , subplot_kw=dict(aspect='equal'))
plt.pie(values , startangle=90 , autopct='%.2f%%')
plt.title('Types of terrorist attack')
plt.legend(labels , loc="upper right" , bbox_to_anchor = (1.3 , 0.9) , fontsize=15)
plt.show()
```

1. In above graph shows that highest attack is armed assult. there is 39.09% of all our attck.
2. and the second largest attack on BOMBING AND EXPLOSION.
3. and less attack on infrastuture attack.

```python
country_data = data.loc[: , 'country':'country_txt']

country_killdata = pd.concat([country_data , killData] , axis=1)
```

```python
country_kill_format_data = country_killdata.pivot_table(columns='country_txt' , values='nkill' , aggfunc='sum')
country_kill_format_data
```

```python
country_kill_format_data.info()
```

```python
fig_size = plt.rcParams['figure.figsize']
fig_size[0]=25
fig_size[1]=25
plt.rcParams['figure.figsize'] = fig_size 
```

```python
labels = country_kill_format_data.columns.tolist()
labels = labels[:50]
index = np.arange(len(labels))
transpose = country_kill_format_data.T
values = transpose.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values]
colors = ['red' , 'blue' , 'green' , 'grey' , 'black' , 'brown' ,'magenta' , 'orange']
fig , ax = plt.subplots(1,1)
ax.yaxis.grid(True)
fig_size = plt.rcParams['figure.figsize']
fig_size[0]=25
fig_size[1]=25
plt.rcParams['figure.figsize'] = fig_size
plt.bar(index , values , color=colors , width=0.9)
plt.ylabel('killes peaple' , fontsize=15)
plt.xticks(index , labels , fontsize=14 , rotation=90)
plt.title('Number of people killed by countries')
plt.show()
```
